```python
import tensorflow as tf
from tensorflow.keras import layers, models
import matplotlib.pyplot as plt
import numpy as np

# -----------------------------
# 1. LOAD DATASET
# -----------------------------
TRAIN_PATH = "/kaggle/input/datasets/sunilthite/cat-or-dog-image-classification"
TEST_PATH  = "/kaggle/input/datasets/sunilthite/cat-or-dog-image-classification"

train_ds = tf.keras.utils.image_dataset_from_directory(
    TRAIN_PATH,
    image_size=(224, 224),
    batch_size=32
)

test_ds = tf.keras.utils.image_dataset_from_directory(
    TEST_PATH,
    image_size=(224, 224),
    batch_size=32
)

class_names = train_ds.class_names
print("Classes:", class_names)

# -----------------------------
# 2. BUILD MODEL
# -----------------------------
num_classes = len(class_names)

cnn_model = models.Sequential([

    # Block 1
    layers.Conv2D(32, (4,4), strides=2, activation='relu', input_shape=(224,224,3)),
    layers.MaxPooling2D(pool_size=(2,2)),

    # Block 2
    layers.Conv2D(64, (4,4), strides=2, activation='relu'),
    layers.MaxPooling2D(pool_size=(2,2)),

    # Block 3
    layers.Conv2D(128, (4,4), strides=2, activation='relu'),
    layers.MaxPooling2D(pool_size=(2,2)),

    # Flatten
    layers.Flatten(),

    # Dense
    layers.Dense(50, activation='relu'),

    # Output
    layers.Dense(num_classes, activation='softmax')
])

# -----------------------------
# 3. COMPILE
# -----------------------------
cnn_model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# -----------------------------
# 4. TRAIN
# -----------------------------
history = cnn_model.fit(
    train_ds,
    validation_data=test_ds,
    epochs=15
)

# -----------------------------
# 5. PREDICTION (SAMPLE)
# -----------------------------
for images, labels in test_ds.take(1):

    sample_images = tf.gather(images, [2, 4])
    sample_labels = tf.gather(labels, [2, 4])

    preds = cnn_model.predict(sample_images)
    pred_labels = np.argmax(preds, axis=1)

# -----------------------------
# 6. VISUALIZATION
# -----------------------------
plt.figure(figsize=(8,4))

for i in range(2):
    plt.subplot(1,2,i+1)
    plt.imshow(sample_images[i].numpy().astype("uint8"))
    plt.title(
        f"Pred: {class_names[pred_labels[i]]}\n"
        f"Actual: {class_names[sample_labels[i]]}"
    )
    plt.axis("off")

plt.show()
```
